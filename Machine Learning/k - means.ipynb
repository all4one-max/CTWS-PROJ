```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
```

```python
df = pd.read_csv("Iris.csv")
classes = df["Species"]
df.head()
```

```python
df2 = df.iloc[ :, 1:-1]
df2
```

```python
df2.head()
```

```python
data = df2.values.tolist() 
data = np.array(data)
## Shuffle classes and data 
data,classes = shuffle(data,classes) 
## First 135 points are used for training and the rest is used for testing
train_data = data[:135]  
test_data = data[135:]
```

```python
c1 = train_data[0]
c2 = train_data[1]
c3 = train_data[2]

def getDistance(center, point):
    dist = (center[0] - point[0])**2 + (center[1] - point[1])**2 + (center[2] - point[2])**2 + (center[3] - point[3])**2
    return dist

def getNewCentroid(old_centroid, cluster):
    if(len(cluster) == 0):
        return old_centroid
    cluster = np.array(cluster)
    sm1 = sum(cluster[:, 0])
    sm2 = sum(cluster[:, 1])
    sm3 = sum(cluster[:, 2])
    sm4 = sum(cluster[:, 3])
    length = len(cluster)
    newCentroid = [sm1/length, sm2/length, sm3/length, sm4/length]
#     print(newCentroid)
    return newCentroid

itr = 1
while(itr <= 100):
    cluster1 = []
    cluster2 = []
    cluster3 = []
    for i in range(len(train_data)):
        dist1 = getDistance(c1, train_data[i])
        dist2 = getDistance(c2, train_data[i])
        dist3 = getDistance(c3, train_data[i])
        mnD = min(dist1, dist2, dist3)
        if(mnD == dist1):
            cluster1.append(train_data[i])
        elif(mnD == dist2):
            cluster2.append(train_data[i])
        else:
            cluster3.append(train_data[i])
    prev_c1 = c1
    prev_c2 = c2
    prev_c3 = c3
    c1 = getNewCentroid(c1, cluster1)
    c2 = getNewCentroid(c2, cluster2)
    c3 = getNewCentroid(c3, cluster3)
    itr += 1
print(itr)
print("Final Centroid", c1, c2, c3)
```

```python
pred = []
for i in range(len(test_data)):
    dist1 = getDistance(c1, train_data[i])
    dist2 = getDistance(c2, train_data[i])
    dist3 = getDistance(c3, train_data[i])
    mnD = min(dist1, dist2, dist3)
    if(mnD == dist1):
        pred.append(1)
    elif(mnD == dist2):
        pred.append(2)
    else:
        pred.append(3)
print(pred)
```
