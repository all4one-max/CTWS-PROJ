```python
### add row
df2 = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
df2.columns = ['sl','sw','pl','pw','flower_type']
df2.loc[149]=[1,2,3,4,'Iris-versicolor']
df2.tail()
```

```python
column_name=['sl','sw','pl','pw','flower_type']
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=column_name)
df2=iris.copy()
df2.drop([0,1,2,3],inplace=True)
df2.reset_index(inplace = True)
df2.head()
```
